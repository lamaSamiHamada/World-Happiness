```python
def happiness_rate(df,col):
    print(f'###{col}###')
    print(df.groupby(col).happiness.mean())
    sns.scatterplot(x=col ,y='happiness',data=df);
    plt.axhline(df.happiness.mean(),color='black',linestyle='--');
    plt.show()
```

```python
df.sample()
```

```python
for col in ['social_support','life_expectancy','freedom','generosity','corrption']:
    happiness_rate(df,col)
```

```python
def outliers(df,col):
    print(f'####{col}###')
    sns.boxplot(x=col,data=df)
    plt.show()
```

```python
for col in ['country','reginol','happiness','gdp','social_support','life_expectancy','freedom','generosity','corrption']:
    outliers(df,col)
```

```python
df=df[df.corrption>=0.5]
df=df[df.generosity<0.4]
df=df[df.freedom>=0.5]
outliers(df,'corrption')
outliers(df,'generosity')
outliers(df,'freedom')
```
